```
from tabulate import tabulate
import random
import json

diccionario = {"Matricula" : "", "Edad" : "", "Nombre" : "",
               "Ap paterno" : "", "Ap materno" : "", "Sexo" : ""}

datos = []

ApP = ["Cardenas", "Meza", "Rojas", "Navarro", "Zavala", "Cazares", "Gutierrez", "Zepeda",
       "Perez", "Sanchez", "Torres", "Casas", "Leal", "Patron", "Martinez"]

ApM = ["Quiroz", "Tamayo", "Rosales", "Blanco", "Juarez", "Encinas", "Aldama", "Saucedo", "Vera",
       "Arroyo", "Marrero", "Gallegos", "Siqueiros", "Guerrero", "Garcia"]

sexo = ["Hombre", "Mujer"]

NombresH = ["Christian", "Inti", "Pavel", "Javier", "Roberto", "Angel", "Daniel", "Julio", "Erick",
            "David", "Jonathan", "Johan", "Benjamin", "Jesus", "Antonio"]

NombresM = ["Cellinia", "Jacqueline", "Maria", "Margareth", "Zofia", "Janneth", "Thamara", "Melissa", "Brisa",
            "Mirsha", "Adle", "Noe", "Justine", "Fernanda", "Roberta"]

##################################################################################
def aleatorio():
  i = 1

  while i <= 10:
    genero = random.choice(sexo)

    if genero == "Mujer":
      datos.append({"Matricula" : random.randrange(300000, 399999), "Edad" : random.randrange(18, 60),
                    "Nombre": random.choice(NombresM), "Ap paterno" : random.choice(ApP), "Ap materno" : random.choice(ApM), "Sexo" : "Mujer"})
      
    else:
      datos.append({"Matricula" : random.randrange(300000, 399999), "Edad" : random.randrange(18, 60),
                    "Nombre": random.choice(NombresH), "Ap paterno" : random.choice(ApP), "Ap materno" : random.choice(ApM), "Sexo" : "Hombre"})
      
    i = i + 1

  print("\n-Datos generados-\n")

##################################################################################
def manual():
  i = 1

  while i <= 1:
    try:
      gender = int(input("1. Hombre | 2. Mujer\n"))
      paterno = str(input("Apellido paterno: "))
      materno = str(input("Apellido materno: "))
      nombre = str(input("Nombre(s): "))
      age = int(input("Edad: "))
      matricula = int(input("Matricula: "))

    except:
      print("\n-Error, vuelve a intentarlo-\n")

    else:

      if gender == 1:
        sex = "Hombre"

      else:
        sex = "Mujer"

      datos.append({"Matricula" : matricula, "Edad" : age, "Nombre" : nombre,
                    "Ap paterno" : paterno, "Ap materno" : materno, "Sexo" : sex})
      
      i = i + 1

  print("\n-Datos agregados-\n")

##################################################################################
def impresion1():
  print("----------------------------------------------------------------------")
  print("Matricula" " " "Edad" " " "Nombre" " " "Ap paterno" " " "Ap materno" " " "Sexo" " " "\n")
  print(tabulate(datos))

##################################################################################
def impresion2():
  print("----------------------------------------------------------------------")
  print(json.dumps(datos, sort_keys=False, indent=2))

##################################################################################
def impresion3():
   print("----------------------------------------------------------------------")
   print("\n". join(map(str, datos)))

##################################################################################
def impresiones():
  op = random.randint(0, 2)

  if op == 0:
    impresion1()

  if op == 1:
    impresion2()

  if op == 2:
    impresion3()

##################################################################################
def buscar():
  band = 0
  i = 1

  try:
    search = int(input("Matricula a buscar: "))

  except:
    print("\n-Error, vuelve a intentarlo-\n")

  else:

    for i in datos:
      if search == i["Matricula"]:
        print("--------------------Datos encontrados!--------------------")
        print("Matricula: ", i["Matricula"])
        print("Nombre: ", i["Nombre"])
        print("Ap paterno: ", i["Ap paterno"])
        print("Ap materno: ", i["Ap materno"])
        print("Edad: ", i["Edad"])
        print("Sexo: ", i["Sexo"])

        band = 1

    if band == 0:
      print("\n-No existe...-\n")

##################################################################################
def ordenar():
  i = 1
  datos.sort(key = lambda i: i["Matricula"])

  print("\n-Ordenando datos...-\n")

##################################################################################
def eliminar():
  i = 1

  try:
    delete = int(input("Matricula a borrar: "))

  except:
    print("\n-Error, vuelve a intentarlo-\n")

  else:

    for i in datos:
      if delete == i["Matricula"]:
        datos.pop(0)

  print("\n-Datos eliminados-\n")

##################################################################################
def deshacer():
  datos.clear()

  print("\n-Se eliminaron todos los datos!-\n")

##################################################################################
def menu():
  while(True):
    print("--------------------Menu--------------------")
    print("1. Generar aleatoriamente")
    print("2. Generar manualmente")
    print("3. Imprimir")
    print("4. Buscar")
    print("5. Ordenar")
    print("6. Eliminar")
    print("7. Eliminar toda la lista")
    print("0. Salir")

    print("--------------------------------------------")
    try:
      op = int(input("Escoge una opcion: "))

    except:
      print("\n-Error, vuelve a intentarlo-\n")

    else:

      if op == 1:
        print("\n")
        aleatorio()

      if op == 2:
        print("\n")
        manual()

      if op == 3:
        impresiones()

      if op == 4:
        print("\n")
        buscar()

      if op == 5:
        print("\n")
        ordenar()

      if op == 6:
        eliminar()

      if op == 7:
        print("\n")
        deshacer()

      if op == 0:
        break

menu()
```
